```python
import pandas as pd
cols = [ 'name' , 'gender', 'birth']

first_task = pd.read_table('yob1880.txt', sep = ',', engine = 'python' , names = cols)
gender_birth = pd.pivot_table(first_task,
               index=['gender'],
              values=['birth'], aggfunc='sum')
print(gender_birth)
```

```python
years = range(1880, 2011)
pieces = []
for i in years:
    df = pd.read_table('yob%d.txt'%i, sep = ',', engine = 'python' , names = cols)
    df['year'] = i
    pieces.append(df)
    data = pd.concat(pieces, ignore_index = True)
print(data)
```

```python
data3 = pd.pivot_table(data, index=['gender', 'year'], values=['birth'], aggfunc='sum')
print(data3['F':'F'].plot(y='birth', use_index=True))
print(data3['M':'M'].plot(y='birth', use_index=True))

```

```python
data4 = data.copy()

data4['proportion'] = data4['birth']/data4['birth'].sum()
print(data4)
```

```python
data5 = pd.pivot_table(data, index=['name', 'year'], values=['birth'], aggfunc='sum')
print(data5['Bob':'Bob'].plot(y='birth', use_index=True))
print(data5['Johnny':'Johnny'].plot(y='birth', use_index=True))
print(data5['Natalie':'Natalie'].plot(y='birth', use_index=True))
print(data5['Danil':'Danil'].plot(y='birth', use_index=True))
```

```python
new_sad = []
for i in years:
    data6 = pd.read_table('yob%d.txt'%i, sep = ',', engine = 'python' , names = cols)
    data6['year'] = i
    data6.sort_values(by='birth', ascending=False)
    data6 = data6.head(1)
    new_sad.append(data6)
    data6 = pd.concat(new_sad, ignore_index = True)
data6
```
